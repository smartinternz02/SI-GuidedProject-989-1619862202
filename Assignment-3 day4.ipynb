```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
iris
```

```python
plt.scatter(iris["sepal_length"],iris["petal_length"])
```

```python
sns.lineplot(x="sepal_width",y="petal_width",data=iris)
```

```python
sns.distplot(iris["petal_length"])
```

```python
sns.barplot(x="species",y="sepal_width",data=iris)
```

```python
sns.boxplot(x="petal_length",y="species",data=iris)
```

```python
sns.catplot(x="species",y="petal_width",data=iris)
```

```python
sns.pairplot(iris)
```

```python
sns.heatmap(iris.corr(),annot=True)
```
